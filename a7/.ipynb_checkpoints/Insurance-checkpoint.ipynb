# Preprocessing

## Dealing with nan values

We see that only join_date have nan values and we fill it with the mode

```python
mode = train["join_date"].mode()
train = train.fillna(str(mode[0]))
test = test.fillna(str(mode[0]))
train['join_date'] = train['join_date'].apply(lambda x: (int(x.split('/')[2]) + int(x.split('/')[1])/12 + int(x.split('/')[0])/365 ) if (x == x) else np.nan)
test['join_date'] = test['join_date'].apply(lambda x: (int(x.split('/')[2]) + int(x.split('/')[1])/12 + int(x.split('/')[0])/365 ) if (x == x) else np.nan)
train['join_date'] = train['join_date'].astype('float')
test['join_date'] = test['join_date'].astype('float')
train
```

# Reformat data

We are going to reformat the data to look like a D single-classifications where D is the number of features and create a new column to stock the target

```python
X_train = []
X_train_columns = train.columns
c = 0
for v in train.values:
    info = v[:8]
    binary = v[8:]
    index = [k for k, i in enumerate(binary) if i == 1]
    for i in index:
        c+=1
        for k in range(len(binary)):
            if k == i:
                binary_transformed = list(copy.copy(binary))
                binary_transformed[i] = 0
                X_train.append(list(info) + binary_transformed + [X_train_columns[8+k]] + [c])

X_train = pd.DataFrame(X_train)
X_train.columns = ['ID', 'join_date', 'sex', 'marital_status', 'birth_year', 'branch_code',
       'occupation_code', 'occupation_category_code', 'P5DA', 'RIBP', '8NN1',
       '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO',
       'BSTQ', 'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3', 'target', 'ID2']

X_train
```

```python
X_train.head(10)
```

```python
X_test = []
true_values = []
c = 0
for v in test.values:
    c += 1
    info = v[:8]
    binary = v[8:]
    index = [k for k, i in enumerate(binary) if i == 1]
    X_test.append(list(info) + list(binary) + [c])
    for k in test.columns[8:][index]:
        true_values.append(v[0] + ' X ' + k)

X_test = pd.DataFrame(X_test)
X_test.columns = ['ID', 'join_date', 'sex', 'marital_status', 'birth_year', 'branch_code',
       'occupation_code', 'occupation_category_code', 'P5DA', 'RIBP', '8NN1',
       '7POT', '66FJ', 'GYSR', 'SOP4', 'RVSZ', 'PYUQ', 'LJR9', 'N2MW', 'AHXO',
       'BSTQ', 'FM3X', 'K6QO', 'QBOL', 'JWFN', 'JZ9D', 'J9JW', 'GHYX', 'ECY3', 'ID2']
X_test
```

## Separate the y_train

```python
y_train = X_train[['target']]
X_train.drop("target", axis = 1, inplace= True)
y_train
```

## Encodage

We use label encoder beacause one hot encoder is very heavy(the values of occupation code is 213)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

# We goup our train and test sets to encode them together
data = X_train.append(X_test)
data.head(5)
```

### We are going to use label encoder to encode the object type of our all dataset

```python
for col in ['sex', 'marital_status', 'branch_code', 'occupation_code', 'occupation_category_code',]:
    data.loc[:,col] = encoder.fit_transform(data.loc[:,col])
X_train = data[:X_train.shape[0]]
X_test = data[-X_test.shape[0]:]
X_train
```

### And we encode the y_train as well

```python
encoder.fit(y_train.iloc[:,0])
y_train = pd.DataFrame(encoder.transform(y_train.iloc[:,0]))
y_train.columns = ['target']
y_train
```

# Model selection

```python
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
```

```python
def predict(model,X_test):
    #Proba prediction
    proba = model.predict_proba(X_test.drop(columns=['ID','ID2'], axis=1))
    y_test = pd.DataFrame(proba)

    # Reverse the code of the product
    y_test.columns = encoder.inverse_transform(y_test.columns)
    
    return y_test
```

```python
# Function adding the proba to the result tab
def generateResult(X_test, y_test):
    result = []
    for i in range(X_test.shape[0]):
        id = X_test['ID'].iloc[i]
        for c in y_test.columns:
            result.append([id + ' X ' + c, y_test[c].iloc[i]])

    # Setting the new column as wanted in the submission example
    df_result = pd.DataFrame(result)
    df_result.columns = ['ID X PCODE', 'Label']
    for i in range(df_result.shape[0]):
        if df_result['ID X PCODE'].iloc[i] in true_values:
            df_result['Label'].iloc[i] = 1.0
    return df_result
```

```python
X_test['birth_year'] = X_test['birth_year'].astype('float') 
```

### XGBoost model

```python
model = XGBClassifier(verbose = 2)
model.fit(X_train.drop(columns=['ID', 'ID2']), y_train)
```

```python
y_test = predict(model,X_test)
```

### Submit XGBoost  prediction

```python
df_result = generateResult(X_test, y_test)
df_result.reset_index(drop=True, inplace=True)
df_result.to_csv('finalXGB.csv', index=False)
```
